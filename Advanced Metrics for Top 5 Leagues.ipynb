```python
# Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Shooting data from FBREF both for player level and team level for Top 5 leagues
shooting = pd.read_csv(r'C:\Users\User\Desktop\Sports Analytics\Soccer\Advanced Metrics\Shooting\Shooting.csv')
shooting_team = pd.read_csv(r'C:\Users\User\Desktop\Sports Analytics\Soccer\Advanced Metrics\Shooting\Shooting_Team.csv')

# Passing data from FBREF both for player level and team level for Top 5 leagues
passing = pd.read_csv(r'C:\Users\User\Desktop\Sports Analytics\Soccer\Advanced Metrics\Passing\Passing.csv')
passing_team = pd.read_csv(r'C:\Users\User\Desktop\Sports Analytics\Soccer\Advanced Metrics\Passing\Passing_Team.csv')

# Goal and Shot Creation data from FBREF both for player level and team level for Top 5 leagues
gca = pd.read_csv(r'C:\Users\User\Desktop\Sports Analytics\Soccer\Advanced Metrics\Shot and Goal Creation\GCA.csv')
gca_team = pd.read_csv(r'C:\Users\User\Desktop\Sports Analytics\Soccer\Advanced Metrics\Shot and Goal Creation\GCA_Team.csv')
```

```python
# Merging the passing and shooting dataframes by escaping duplicate columns
team_Stats = pd.merge(shooting_team, passing_team, how='outer', suffixes=('', '_remove'),on = ['Squad'])
  
# remove the duplicate columns
team_Stats.drop([i for i in team_Stats.columns if 'remove' in i],
               axis=1, inplace=True)

# Merging the passing and shooting dataframes by escaping duplicate columns
team_Stats = pd.merge(team_Stats, gca_team, how='outer', suffixes=('', '_remove'),on = ['Squad'])
  
# remove the duplicate columns
team_Stats.drop([i for i in team_Stats.columns if 'remove' in i],
               axis=1, inplace=True)

team_Stats.sort_values(by = "SCA90_SCA",ascending = False)
```

```python
# Discarding the opponent team's statistics from the dataframe
team_Stats = team_Stats[team_Stats.Team_or_Opponent == "team"]

# Replacing the season notation
team_Stats['Season_End_Year'] = team_Stats['Season_End_Year'].replace(['2022'],'2021-2022')

# Listing columns of team level shooting data
team_Stats.columns
```

```python
# Check whether there is duplicate teams in the dataframe
print(team_Stats.Squad.value_counts())

# Dropping duplicate teams from the dataframe
team_Stats.drop_duplicates(subset=['Squad'],inplace = True)
```

```python
# Rather than working with total numbers, P90 metrics are obtained (Especially makes sense in the player level data)
team_Stats['Gls_p90'] = team_Stats['Gls_Standard'] / team_Stats['Mins_Per_90']
team_Stats['xG_p90'] = team_Stats['xG_Expected'] / team_Stats['Mins_Per_90']
team_Stats['Ast_p90'] = team_Stats['Ast'] / team_Stats['Mins_Per_90']
team_Stats['xA_p90'] = team_Stats['xA'] / team_Stats['Mins_Per_90']
```

```python
# Scatter plot of goals scored vs expected goal per 90 min colored based on the league
b = sns.scatterplot(team_Stats.Gls_p90,team_Stats.xG_p90, hue=team_Stats.Comp)
```

```python
# Scatter plot of goals scored vs expected goal per 90 min labeled with the team's name
def label_point(x, y, val, ax):
    a = pd.concat({'x': x, 'y': y, 'val': val}, axis=1)
    for i, point in a.iterrows():
        ax.text(point['x']+.02, point['y'], str(point['val']))

ax = team_Stats.plot.scatter('Gls_p90', 'xG_p90',figsize=(20,15))
label_point(team_Stats['Gls_p90'], team_Stats['xG_p90'], team_Stats['Squad'], ax)
```
